```python
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
df = pd.read_csv('post-election-violence-data.csv')
df.drop(0, inplace=True)
```

```python
df['event_date'] = pd.to_datetime(df['event_date'])
df.index = df['event_date']
```

```python
df['fatalities'] =df['fatalities'].apply(lambda num: int(num))
```

```python
df2007 =df.loc['2007-12-31':'2007-01-01']
```

```python
df2007.groupby(df2007['event_type'])['fatalities'].sum()
```

```python
df2007.groupby(df2007['event_type'])['fatalities'].sum().plot.pie(autopct='%1.0f%%', figsize=(14,10), fontsize=16) 
title = '2007 - Events with most fatalities'
plt.title(title, fontsize=20)
plt.ylabel(' ')
#Save Image
plt.savefig(title, dpi=1000, facecolor='w')
```

```python
df2007.groupby([df2007['event_type'], df2007['sub_event_type']])['fatalities'].sum()
```

```python
df2007.groupby([df2007['event_type'], df2007['sub_event_type']])['fatalities'].sum().sort_values(ascending=False)[:4].plot.pie(autopct='%1.0f%%',figsize=(14,10), fontsize=16) 
title = '2007 - Leading causes for violence'
plt.title(title, fontsize=20)
plt.ylabel(' ')
#Save Image
plt.savefig(title, dpi=1000, facecolor='w')
```

```python
df2007.groupby(df2007['admin1'])['fatalities'].sum().sort_values(ascending=False)[:20].plot(kind='barh', figsize=(14,10), fontsize=16)
title = '2007 - How various Regions were affected'
plt.title(title, fontsize=20)
plt.ylabel('Regions',fontsize=14)
plt.xlabel('Number of fatalities',fontsize=14)
#Save Image
plt.savefig(title, dpi=1000, facecolor='w')
```

```python
regions = df2007.groupby([df2007['admin1'],df2007['sub_event_type']])['fatalities'].sum()
```

```python
regions[['Nairobi', 'Kisumu', 'Uasin Gishu']].plot(kind='barh', figsize=(14,10),fontsize=16)
title = '2007 - Sub events in regions'
plt.title(title, fontsize=20)
plt.ylabel('Regions - Sub event', fontsize=14)
plt.xlabel('Number of fatalities', fontsize=14)
#Save Image
plt.savefig(title, dpi=1000, facecolor='w')
```
