```python
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
df = pd.read_csv('post-election-violence-data.csv')
df.drop(0, inplace=True)
```

```python
df['event_date'] = pd.to_datetime(df['event_date'])
df.index = df['event_date']
```

```python
df['fatalities'] =df['fatalities'].apply(lambda num: int(num))
```

```python
df2008 =df.loc['2008-12-31':'2008-01-01']
```

```python
df2008.groupby(df2008['event_type'])['fatalities'].sum()
```

```python
df2008.groupby(df2008['event_type'])['fatalities'].sum().plot.pie(autopct='%1.0f%%', figsize=(14,10), fontsize=16) 
title = '2008 - Events with most fatalities'
plt.title(title, fontsize=20)
plt.ylabel(' ')
#Save Image
plt.savefig(title, dpi=1000, facecolor='w')
```

```python
df2008.groupby([df2008['event_type'], df2008['sub_event_type']])['fatalities'].sum()
```

```python
df2008.groupby([df2008['event_type'], df2008['sub_event_type']])['fatalities'].sum().sort_values(ascending=False)[:5].plot.pie(autopct='%1.0f%%',figsize=(14,10), fontsize=16) 
title = '2008 - Leading causes for violence'
plt.title(title, fontsize=20)
plt.ylabel(' ')
#Save Image
plt.savefig(title, dpi=1000, facecolor='w')
```

```python
df2008.groupby(df2008['admin1'])['fatalities'].sum().sort_values(ascending=False)[:20].plot(kind='barh', figsize=(14,10), fontsize=16)
title = '2008 - How various Regions were affected'
plt.title(title, fontsize=20)
plt.ylabel('Regions',fontsize=14)
plt.xlabel('Number of fatalities',fontsize=14)
#Save Image
plt.savefig(title, dpi=1000, facecolor='w')
```

```python
regions = df2008.groupby([df2008['admin1'],df2008['sub_event_type']])['fatalities'].sum()
```

```python
regions[['Turkana', 'Nakuru', 'Trans Nzoia']].plot(kind='barh', figsize=(14,10),fontsize=16)
title = '2008 - Sub events in regions'
plt.title(title, fontsize=20)
plt.ylabel('Regions - Sub event', fontsize=14)
plt.xlabel('Number of fatalities', fontsize=14)
#Save Image
plt.savefig(title, dpi=1000, facecolor='w')
```
