```python
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
df = pd.read_csv('post-election-violence-data.csv')
df.drop(0, inplace=True)
```

```python
df['event_date'] = pd.to_datetime(df['event_date'])
df.index = df['event_date']
```

```python
df['fatalities'] =df['fatalities'].apply(lambda num: int(num))
```

```python
df2017 =df.loc['2017-12-31':'2017-01-01']
```

```python
df2017.groupby(df2017['event_type'])['fatalities'].sum()
```

```python
df2017.groupby(df2017['event_type'])['fatalities'].sum().plot.pie(autopct='%1.0f%%', figsize=(14,10), fontsize=16) 
title = '2017 - Events with most fatalities'
plt.title(title, fontsize=20)
plt.ylabel(' ')
#Save Image
plt.savefig(title, dpi=1000, facecolor='w')
```

```python
df2017.groupby([df2017['event_type'], df2017['sub_event_type']])['fatalities'].sum()
```

```python
df2017.groupby([df2017['event_type'], df2017['sub_event_type']])['fatalities'].sum().sort_values(ascending=False)[:7].plot.pie(autopct='%1.0f%%',figsize=(14,10), fontsize=16) 
title = '2017 - Leading causes for violence'
plt.title(title, fontsize=20)
plt.ylabel(' ')
#Save Image
plt.savefig(title, dpi=1000, facecolor='w')
```

```python
df2017.groupby(df2017['admin1'])['fatalities'].sum().sort_values(ascending=False)[:20].plot(kind='barh', figsize=(14,10), fontsize=16)
title = '2017 - How various Regions were affected'
plt.title(title, fontsize=20)
plt.ylabel('Regions',fontsize=14)
plt.xlabel('Number of fatalities',fontsize=14)
#Save Image
plt.savefig(title, dpi=1000, facecolor='w')
```

```python
regions = df2017.groupby([df2017['admin1'],df2017['sub_event_type']])['fatalities'].sum()
```

```python
regions[['Nairobi', 'Lamu', 'Mandera']].plot(kind='barh', figsize=(14,10),fontsize=16)
title = '2017 - Sub events in regions'
plt.title(title, fontsize=20)
plt.ylabel('Regions - Sub event', fontsize=14)
plt.xlabel('Number of fatalities', fontsize=14)
#Save Image
plt.savefig(title, dpi=1000, facecolor='w')
```
