```python
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
df = pd.read_csv('post-election-violence-data.csv')
df.drop(0, inplace=True)
```

```python
df['event_date'] = pd.to_datetime(df['event_date'])
df.index = df['event_date']
```

```python
df['fatalities'] =df['fatalities'].apply(lambda num: int(num))
```

```python
df2013 =df.loc['2013-12-31':'2013-01-01']
```

```python
df2013.groupby(df2013['event_type'])['fatalities'].sum()
```

```python
df2013.groupby(df2013['event_type'])['fatalities'].sum().plot(kind='pie',figsize=(14,10))
plt.title(' 2013 - Events with most fatalities')
plt.ylabel(' ')
```

```python
df2013.groupby([df2013['event_type'], df2013['sub_event_type']])['fatalities'].sum()
```

```python
df2013.groupby([df2013['event_type'], df2013['sub_event_type']])['fatalities'].sum().sort_values(ascending=False).plot(kind='pie', figsize=(14,10))
plt.title(' 2013 - Leading causes for violence')
plt.ylabel(' ')
```

```python
df2013.groupby(df2013['admin1'])['fatalities'].sum().sort_values(ascending=False)[:20].plot(kind='barh', figsize=(14,10))
plt.title(' 2013 - How various Regions were affected')
plt.ylabel('Regions')
plt.xlabel('Number of fatalities')
```

```python
regions = df2013.groupby([df2013['admin1'],df2013['sub_event_type']])['fatalities'].sum()
```

```python
regions[['Mandera', 'Nairobi', 'Garissa']].plot(kind='barh', figsize=(14,10))
plt.title('2013 - Sub events in regions')
plt.ylabel('Regions - Sub event')
plt.xlabel('Number of fatalities')
```
